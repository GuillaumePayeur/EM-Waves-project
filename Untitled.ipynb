Plotting evolution of energy density in magnetic field during inflation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200
plt.rcParams.update({"text.usetex": True})
from scipy.interpolate import interp1d
from numpy.lib.scimath import sqrt as csqrt
```

```python
from scipy.special import gamma
```

```python
def F_1(alpha):
    return np.pi/(2**(1-2*alpha)*(gamma((1/2)-alpha))**2*(np.cos(-np.pi*alpha))**2)
def F_2(alpha):
    return np.pi/(2**(2*alpha+3)*(gamma((3/2)+alpha))**2*(np.cos(np.pi*(alpha+1)))**2)
```

```python
alpha_array_1 = np.linspace(-1/2,2,500)
alpha_array_2= np.linspace(-3,-1/2,500)
F_1_array = F_1(alpha_array_1)
F_2_array = F_2(alpha_array_2)

alpha_array = np.concatenate((alpha_array_2,alpha_array_1))
F_array = np.concatenate((F_2_array,F_1_array))
plt.plot(alpha_array,F_array,color='green')
plt.plot([-1/2,-1/2],[0,4],ls='--',color='gray',label='$\\alpha=-1/2$')
plt.ylim(0,4)
plt.xlabel('$\\alpha$',fontsize=15)
plt.ylabel('$\mathcal{F}(\\alpha)$',fontsize=15)
plt.legend(loc=2,frameon=True,fontsize=10)
plt.tight_layout()
plt.savefig('F.pdf')
```

```python
def G_1(alpha):
    return np.pi/(2**(-2*alpha+5)*(gamma(-alpha+5/2))**2*(np.cos(np.pi*(-alpha+1)))**2)
def G_2(alpha):
    return np.pi/(2**(2*alpha+3)*(gamma((3/2)+alpha))**2*(np.cos(np.pi*(alpha)))**2)
```

```python
alpha_array_1 = np.linspace(1/2,4,500)
alpha_array_2= np.linspace(-3,1/2,500)
G_1_array = G_1(alpha_array_1)
G_2_array = G_2(alpha_array_2)

alpha_array = np.concatenate((alpha_array_2,alpha_array_1))
G_array = np.concatenate((G_2_array,G_1_array))
plt.plot(alpha_array,G_array,color='red')
plt.plot([-1/2,-1/2],[0,4],ls='--',color='gray',label='$\\alpha=-1/2$')
plt.plot([1/2,1/2],[0,4],ls='--',color='gray',label='$\\alpha=1/2$')
plt.plot([3/2,3/2],[0,4],ls='--',color='gray',label='$\\alpha=3/2$')
plt.ylim(0,4)
plt.xlabel('$\\alpha$',fontsize=15)
plt.ylabel('$\mathcal{G}(\\alpha)$',fontsize=15)
plt.legend(loc=2,frameon=True,fontsize=10)
plt.tight_layout()
plt.savefig('G.pdf')
```
